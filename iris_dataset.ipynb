#Model Building.

```python
#data variable
x = nds.drop('Species', axis=1)
print(x.shape)

#target variable
y = nds['Species']
print(y.shape)
```

```python
#splitting the training and testing datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
```

*   x_train has the training features.
*   x_test has the testing features.
*   y_train has the training labels.
*   y_test has the testing labels.





##ALGORITHMS

```python
##DECISION TREE
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
#this creates an empty model

#fitting the empty model
classifier.fit(x_train, y_train)

#making predictions
predictions = classifier.predict(x_test)

```

```python
#matching predictions with expected outcome to measure the accuracy value.
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
##K-NEAREST NEIGHBORS
#creating an empty model
from sklearn import neighbors
classifier= neighbors.KNeighborsClassifier()

#fitting the model
classifier.fit(x_train, y_train)

#make predictions
predictions = classifier.predict(x_test)
```

```python
#matching predictions with expected outcome to measure the accuracy value.
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
##LOGISTIC REGRESSION
#create empty model
from sklearn.linear_model import LogisticRegression
classifier_LR = LogisticRegression()

#fit the model
classifier_LR.fit(x_train, y_train)

#make predictions
predict_LR =classifier_LR.predict(x_test)
```

```python
#matching predictions with expected outcome to measure the accuracy value.
#from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict_LR)
```

```python
#choosing the Logistic Regression since it has the higest accuracy score
classifier_LR.fit(x, y)

# make a prediction for an example of an out-of-sample observation
classifier_LR.predict([[6, 3, 4, 2]])
```
